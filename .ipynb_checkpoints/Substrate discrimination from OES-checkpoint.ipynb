```python
## TEMPERATURE ESTIMATION from OES
## BASED ON N2(CB) TRANSITION
## Author Dogan Gidon, 2018/07/12

#import tools
import numpy as np
import matplotlib.pyplot as pl
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
#from scipy import io as sio
import pandas as pd

################################################################################
# LOAD TRAINING AND TEST DATA
# first 51 columns are the OES signal
# last two columns are rotational temeprature and vibrational temperature

df_train=pd.read_csv('dat_train.csv') ##import training data
df_test=pd.read_csv('dat_test.csv') ##import test data
df_test_p=pd.read_csv('dat_test_pull.csv') ##import test data
```

```python
## visualize independent variabiles i.e. OES peaks

import matplotlib.pyplot as plt
l=np.arange(0,len(df_train))
plt.figure()
plt.plot(np.transpose(np.array(df_train.iloc[1:895,0:51])))

plt.show()
```

```python
#select training and test output data
Xtrain=np.array(df_train.iloc[1:895,0:51])
Xtest=np.array(df_test.iloc[:,0:51])
Xpull=np.array(df_test_p.iloc[:,0:51])

XX=np.vstack((Xtrain,Xtest))
#select substate labels as outputs ***for validation only***
Ltrain=df_train.iloc[1:895,55]
Ltest=df_test.iloc[:,55]
Lpull=np.array(df_test_p.iloc[:,55])
```

```python
## K-means procedure
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score

#fit and compare the clusters in the training data to the susbstrate labels
kmeans=KMeans(n_clusters=2, random_state=1)
kmeans.fit(Xtrain) #note substrates labels are NOT used for training

Lpred_train=kmeans.predict(Xtrain)


plt.plot(Lpred_train,'o',label='predicted')
plt.plot(np.hstack((Ltrain)),label='known')
plt.ylabel('Substrate Type')
plt.xlabel('Measurement Instance')
plt.legend()

plt.figure()
plt.plot(df_train.iloc[1:895,51],label='Power')
plt.plot(df_train.iloc[1:895,52],label='Flow')
plt.legend()
plt.ylabel('Input')
plt.xlabel('Measurement Instance')
plt.show()

print('Accuracy Score: {:.3f}%'.format(100.*accuracy_score(Lpred_train, Ltrain)))
```

```python
#compare against test data
Lpred_test=kmeans.predict(Xtest)
plt.plot(Lpred_test,'o',label='predicted')
plt.plot(np.hstack((Ltest)),label='known')
plt.ylabel('Substrate Type')
plt.xlabel('Measurement Instance')
plt.legend()
plt.show()

plt.figure()
plt.plot(df_test.iloc[1:895,51],label='Power')
plt.plot(df_test.iloc[1:895,52],label='Flow')
plt.legend()
plt.ylabel('Input')
plt.xlabel('Measurement Instance')
plt.show()

print('Accuracy Score: {:.3f}%'.format(100.*accuracy_score(Lpred_test, Ltest)))
```

```python
#compare against test data
Lpred_pull=kmeans.predict(Xpull)

plt.plot(Lpred_pull,'o',label='predicted')
plt.plot(np.hstack((Lpull)),label='known')
plt.ylabel('Substrate Type')
plt.xlabel('Measurement Instance')
plt.legend()
plt.show()

plt.figure()
plt.plot(df_test_p.iloc[1:895,51],label='Power')
plt.plot(df_test_p.iloc[1:895,52],label='Flow')
plt.legend()
plt.ylabel('Input')
plt.xlabel('Measurement Instance')
plt.show()

print('Accuracy Score: {:.3f}%'.format(100*accuracy_score(Lpred_pull, Lpull)))
```

```python
#plot centroids of the clusters

centroids = kmeans.cluster_centers_
plt.figure()
plt.plot(centroids.transpose()[:,0],'o--')
plt.plot(centroids.transpose()[:,1],'s--')
plt.legend(['avg spectrum for metal','avg spectrum for glass'])
plt.ylabel('Normalized Intensity (a.u.)')
plt.xlabel('Normalized Wavenumber (a.u.)')


plt.show()
```
